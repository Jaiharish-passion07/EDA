# Remove Outliers using IQR

```python
Q1=air_df_1['NOx(GT)'].quantile(.25)
Q3=air_df_1['NOx(GT)'].quantile(.75)
#print(Q1,Q3)
```

```python
Q_1=air_df_1['PT08.S4(NO2)'].quantile(.25)
Q_3=air_df_1['PT08.S4(NO2)'].quantile(.75)
#print(Q1,Q3)
```

```python
IQR=Q3-Q1
```

```python
IQ_R=Q_3-Q_1
```

```python
lower_limits=Q1-(1.5*IQR)
upper_limits=Q3+(1.5*IQR)
```

```python
lower_limits_1=Q_1-(1.5*IQ_R)
upper_limits_1=Q_3+(1.5*IQ_R)
```

```python
air_df_real_1=air_df_1[(air_df_1['NOx(GT)']>lower_limits) & (air_df_1['NOx(GT)']<upper_limits) & (air_df_1['PT08.S4(NO2)']>lower_limits_1) & (air_df_1['PT08.S4(NO2)']<upper_limits_1) ]
```

```python
plt.figure(figsize=(18,10))
sns.boxplot(data=air_df_real_1,color='#02FF0C')
```

Neglecting NMHC Feeatures

```python
air_df_real_1.drop('NMHC(GT)',axis=1,inplace=True)
```

# Model Implementation

```python
from sklearn.preprocessing import StandardScaler         #import normalisation package
from sklearn.model_selection import train_test_split      #import train test split
```

```python
x=air_df_real_1.drop('RH',axis=1)
y=air_df_real_1['RH']
```

```python
s_sr=StandardScaler()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
s_sr=StandardScaler()
x_scalar_train=s_sr.fit_transform(x_train)
x_scalar_test=s_sr.transform(x_test)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

# Model 1 using Linear regression

```python
from sklearn.linear_model import LinearRegression         #import linear regression package
from sklearn.metrics import mean_squared_error,mean_absolute_error   #import mean squared error and mean absolute error
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_scalar_train,y_train)
```

```python
actual_values=y_test
```

```python
predicted_values=lr.predict(x_scalar_test)
```

```python
def knn_model(x):
    knn=KNeighborsRegressor(n_neighbors=x)
    #knn=KNeighborsRegressor()
    knn.fit(x_scalar_train,y_train)
    #actual_values_1=y_test
    predicted_values1=knn.predict(x_scalar_test)
    result_1=pd.DataFrame({'Absolute':y_test,'Predict':predicted_values1})
    result_1
    from sklearn.metrics import r2_score
    R2_Score=r2_score(y_test,predicted_values1)
    return R2_Score
```

```python
res=[]
for i in range(1,15):
    #print(i,knn_model(i))
    res.append(knn_model(i))
```

```python
x=[i for i in range(1,15)]
plt.plot(x,res,'y')
plt.plot(x,res,'g*')
plt.xlabel('K-Values')
plt.ylabel('Accuracy')
plt.title('Relationship between K-Values and Accuracy')
plt.show()
```
